1.	Given a list [4, 8, 2, 10, 6], find the maximum number using a for loop.


```python
list = [4, 8, 2, 10, 6]
maximum = list[0]
for n in list:
  if n > maximum:
    maximum = n
print(f"The maximum number in list is {maximum}")
```

2.	Take input from the user and find the factorial of the number using a for loop.


```python
num = int(input("Enter a number = "))

factorial = 1
for n in range(1, num+1):
  factorial *= n
print(f"factorial of {num} is {factorial}" )
```

4.	Take a string input from the user and count the number of uppercase letters using a for loop. (Use char.isupper()) ðŸ˜Š


```python
string = input("Enter a string = ")
upper_count = 0

for char in string:
  if char.isupper():
    upper_count += 1
print(f"Count of uppercase is {upper_count} ")
```

10.	Given a list of words ['Python', 'Java', 'C++', 'JavaScript', 'Ruby'], find the longest word using a for loop.


```python
list = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
longest_word = list[0]
for char in list:
  if len(char) > len(longest_word):
    longest_word = char
print(f"longest word is {longest_word}")
```

13.	Given a list [1, 3, 5, 7, 9, 11], check if a user-inputted number is present in the list using a for loop.


```python
list = [1, 3, 5, 7, 9, 11]
user = int(input("Enter a number = "))
found = False
for n in list:
  if n == user:
    found = True
    break
if found:
  print("Number is present in list")
else:
  print("Number is not present in list")
```

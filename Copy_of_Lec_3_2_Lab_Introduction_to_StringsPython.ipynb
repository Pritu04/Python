Q3 Write a Python program that accepts a string from user. Your program should create a new string by shifting one position to left.

For example if the user enters the string 'examination 2021' then new string would be 'xamination 2021e'

```python
def shift_left(i_string):

  if not i_string:
    return ""
  first_char = i_string[0]
  remaining_chars = i_string[1:]
  new_string = remaining_chars + first_char
  return new_string
user_input = input("Enter a string: ")
shifted_string = shift_left(user_input)
print(f"Original string: '{user_input}'")
print(f"New string after left shift: '{shifted_string}'")
```

Q5 Write a Python program to get a string made of the first 2 and last 2 characters of a given string. If the string length is less than 2, return the empty string instead.
Sample String : 'w3resource'
Expected Result : 'w3ce'
Sample String : 'w3'
Expected Result : 'w3w3'
Sample String : ' w'
Expected Result : Empty String

```python
def get_first_and_last_two_chars(s):
  if len(s) < 2:
    return ""
  else:
    return s[:2] + s[-2:]

# Sample usage:
print(f"'w3resource' -> '{get_first_and_last_two_chars('w3resource')}'")
print(f"'w3' -> '{get_first_and_last_two_chars('w3')}'")
print(f"' w' -> '{get_first_and_last_two_chars(' w')}'")
print(f"'a' -> '{get_first_and_last_two_chars('a')}'")
print(f"'' -> '{get_first_and_last_two_chars('')}'")
```

Q6 Write a Python program to add 'ing' at the end of a given string (length should be at least 3). If the given string already ends with 'ing', add 'ly' instead. If the string length of the given string is less than 3, leave it unchanged.
Sample String : 'abc'
Expected Result : 'abcing'
Sample String : 'string'
Expected Result : 'stringly'

```python
def modify_string(s):
  if len(s) < 3:
    return s
  elif s.endswith('ing'):
    return s + 'ly'
  else:
    return s + 'ing'

# Sample Usage
print(f"'abc' becomes: {modify_string('abc')}")
print(f"'string' becomes: {modify_string('string')}")
print(f"'go' becomes: {modify_string('go')}")
print(f"'play' becomes: {modify_string('play')}")
```

Q7 Write a Python program to remove the nth index character from a nonempty string.

```python
def remove_nth_char(input_string, n):
  if not input_string:
    return ""
  if n < 0 or n >= len(input_string):
    print(f"Warning: Index {n} is out of bounds for string '{input_string}'. Returning original string.")
    return input_string
  else:
    return input_string[:n] + input_string[n+1:]
my_string = "Hello World"
index_to_remove = 6

modified_string = remove_nth_char(my_string, index_to_remove)
print(f"Original string: '{my_string}'")
print(f"Modified string: '{modified_string}'")

my_string_2 = "Python"
index_to_remove_2 = 0

modified_string_2 = remove_nth_char(my_string_2, index_to_remove_2)
print(f"Original string: '{my_string_2}'")
print(f"Modified string: '{modified_string_2}'")

my_string_3 = "Test"
index_to_remove_3 = 4

modified_string_3 = remove_nth_char(my_string_3, index_to_remove_3)
print(f"Original string: '{my_string_3}'")
print(f"Modified string: '{modified_string_3}'")
```

Q8 Write a Python program to change a given string to a newly string where the first and last chars have been exchanged.

```python
def swap_first_last_chars(input_string):

  if len(input_string) < 2:
    return input_string
  else:
    first_char = input_string[0]
    last_char = input_string[-1]
    middle_part = input_string[1:-1]
    return last_char + middle_part + first_char
string1 = "Python"
new_string1 = swap_first_last_chars(string1)
print(f"Original string: '{string1}'")
print(f"New string: '{new_string1}'")

string2 = "hello"
new_string2 = swap_first_last_chars(string2)
print(f"Original string: '{string2}'")
print(f"New string: '{new_string2}'")

string3 = "a"
new_string3 = swap_first_last_chars(string3)
print(f"Original string: '{string3}'")
print(f"New string: '{new_string3}'")

string4 = ""
new_string4 = swap_first_last_chars(string4)
print(f"Original string: '{string4}'")
print(f"New string: '{new_string4}'")
```

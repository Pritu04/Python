```python
numbers = [2,3,4,5,11,22,33,9,8,10]

#Asc
numbers.sort()
#cube
cube_list = []
for n in numbers:
    cube_list.append(n ** 3)

#  dict with number as key and cube as value
result_dict = dict(zip(numbers, cube_list))

print("Sorted List:", numbers)
print("Cube List:", cube_list)
print("Dictionary:", result_dict)
```

1.Given a list [2, 3, 5, 7, 11, 13, 17, 19], find the index of the first occurrence of the number 7 using a for loop.

```python
numbers = [2, 3, 5, 7, 11, 13, 7, 19]

i= numbers.index(7)
print(i)
```

2.Given a list of mixed values [1, 'hello', 3.5, 'world', 10, 'Python'], print only the integer values using a for loop.

```python
a = [1, 'hello', 3.5, 'world', 10, 'Python']

for i in a:
    if type(i) == int:
        print(i)
```

3.Given a dictionary {'A': 10, 'B': 20, 'C': 30, 'D': 40}, print all key-value pairs using a for loop.

```python
d= {'A': 10, 'B': 20, 'C': 30, 'D': 40}

for key, value in d.items():
    print(key, ":", value)
```
